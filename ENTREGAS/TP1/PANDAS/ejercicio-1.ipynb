**Analisis previo**

* Vemos que en algunas columnas tenemos valores NaN. Analicemos algunos casos que los contengan para ver como manejar esto: 

  "title": Si tiene ID consideramos que es un contenido valido y por lo que vemos ningun valor en "id" es NaN.

  "revisor_comment": Consideramos que un NaN no implica que no se haya hecho una edicion, por lo tanto contabilizamos estas filas.

* Como la cantidad de valores unicos de la columna "id" y de "revision_id" son iguales, quiere decir que para cada id de contenido tenemos únicamente un revision_id. Por lo tanto, son todas las ultimas ediciones.

* Viendo las columnas del df vemos que aquellas que nos pueden interesar para resolver el ejercicio son "revision_timestamp" para calcular lo relacionado a las fechas que nos piden, "revisor_id" y "revisor_ip" para hallar el usuario. Además, con el analisis hecho previamente sabemos que estas columnas no contienen valores NaN (exceptuando en el caso de que un usuario tenga ID no tendra IP y viceversa). Cargamos el archivo unicamente con estas columnas para ahorrarnos el paso de dropear las demás.

```python
df_contenidos = pd.read_csv("/content/drive/MyDrive/Organizacion de Datos/Colab Notebooks/TP1/contents.csv", usecols=["revision_timestamp", "revisor_id", "revisor_ip"])

df_contenidos["revision_timestamp"] = pd.to_datetime(df_contenidos["revision_timestamp"])

df_ediciones_id = df_contenidos.groupby(["revisor_id"]).agg({"revisor_id": ["count"]})
df_ediciones_id.columns = ["cantidad"]
max_cantidad_id = df_ediciones_id["cantidad"].max()
id = df_ediciones_id[df_ediciones_id["cantidad"] >= max_cantidad_id].reset_index()["revisor_id"][0]

df_ediciones_ip = df_contenidos.groupby(["revisor_ip"]).agg({"revisor_ip": ["count"]})
df_ediciones_ip.columns = ["cantidad"]
max_cantidad_ip = df_ediciones_ip["cantidad"].max()
ip = df_ediciones_ip[df_ediciones_ip["cantidad"] >= max_cantidad_ip].reset_index()["revisor_ip"][0]

usuario_buscado = id if (df_ediciones_id.loc[id, "cantidad"] > df_ediciones_ip.loc[ip, "cantidad"]) else ip
df_filtrado = df_contenidos[(df_contenidos["revisor_id"] == usuario_buscado) | (df_contenidos["revisor_ip"] == usuario_buscado)]

fecha_promedio = df_filtrado["revision_timestamp"].mean()
fecha_minima = df_filtrado["revision_timestamp"].min()
fecha_maxima = df_filtrado["revision_timestamp"].max()
```

Respuesta

```python
fecha_promedio, fecha_minima, fecha_maxima
```
