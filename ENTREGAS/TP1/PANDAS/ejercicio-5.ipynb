5) La palabra más común entre los títulos que no sea una stopword del inglés ni español (⭐⭐).

```python
import pandas as pd

from google.colab import drive
drive.mount("/content/drive")
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stopwords = set(stopwords.words('english')).union(stopwords.words('spanish'))
```

**Analisis previo**

* Vemos que en la columna "title" tenemos valores NaN. Como debemos hallar la palabra mas comun entre los titulos debemos considerar eliminar estos valores NaN ya que no nos sirve para nuestro analisis.

* Como la cantidad de valores unicos de la columna "title" y de "id" son iguales, quiere decir que para cada titulo de contenido tenemos únicamente un id. Por lo tanto, no tenemos titulos repetidos.

* Viendo las columnas del df vemos que la unica que nos pueden interesar para resolver el ejercicio es "title" la cual vimos que contiene valores NaN. Cargamos el archivo unicamente con esta columna para ahorrarnos el paso de dropear las demás.

* Una palabra debe tener 2 caracteres como minimo para ser considerada.

```python
df_contenidos = pd.read_csv("/content/drive/MyDrive/Organizacion de Datos/Colab Notebooks/TP1/contents.csv", usecols=["title"])

df_contenidos.dropna(subset=["title"], inplace=True)

df_contenidos["palabras"] = df_contenidos["title"].map( lambda x: x.lower())
df_contenidos["palabras"] = df_contenidos["palabras"].map(lambda x: re.findall(r"\w+(?:'\w+)?|[^\w\s]", x))
df_contenidos["palabras"] = df_contenidos["palabras"].map(lambda x: [palabra for palabra in x if (len(palabra) > 1) and (palabra not in stopwords)])

df_contenidos = df_contenidos.explode("palabras")

cantidad_de_palabras = df_contenidos.groupby("palabras").agg({"palabras": ["count"]})
cantidad_de_palabras.columns = ["apariciones"]
max_cantidad = cantidad_de_palabras["apariciones"].max()
resultado = cantidad_de_palabras[cantidad_de_palabras["apariciones"] >= max_cantidad].reset_index()["palabras"][0]
```

Respuesta

```python
print("La palabra más común entre los títulos que no sea una stopword del inglés ni español es '" + resultado + "'.")
```
