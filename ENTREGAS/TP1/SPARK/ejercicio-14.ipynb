### **Resolucion explicativa**

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext

sqlContext = SQLContext(sc)
df = sqlContext.read.csv('/content/drive/MyDrive/Organizacion de Datos/Colab Notebooks/TP1/redirect_list.csv', header=True, inferSchema=True)
rdd = df.rdd
```

Pasos

1) Con la transformacion map() buscamos obtener registros de la forma (Título del contenido al que redirige, 1) para poder contarlos.

2) Con la transformacion reduceByKey() buscamos sumar los registros para una misma clave.

3) Con la accion takeOrdered() buscamos hallar los primeros 5 registros de mayor a menor.

```python
resultado = rdd.filter(lambda x: x.rd_title != "NaN").map(lambda x: (x.rd_title, 1)).reduceByKey(lambda x,y: x+y).takeOrdered(5, lambda x: -x[1])
```

### **Resolucion reducida**

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext

sqlContext = SQLContext(sc)
df = sqlContext.read.csv('/content/drive/MyDrive/Organizacion de Datos/Colab Notebooks/TP1/redirect_list.csv', header=True, inferSchema=True)
rdd = df.rdd

resultado = rdd.filter(lambda x: x.rd_title != "NaN").map(lambda x: (x.rd_title, 1)).reduceByKey(lambda x,y: x+y).takeOrdered(5, lambda x: -x[1])
```
