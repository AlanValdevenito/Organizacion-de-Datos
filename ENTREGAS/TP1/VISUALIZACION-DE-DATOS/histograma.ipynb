```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount("/content/drive")
```

**Analisis previo**

* Notemos que no tenemos ningun valor NaN en las columnas que nos interesan.

```python
df = pd.read_csv("/content/drive/MyDrive/Organizacion de Datos/Trabajos Practicos/soja.csv", usecols = ["Fecha", "Cierre"])

df["Año"] = df["Fecha"].map(lambda x: x.split(" ")[0].split("/")[2]).astype(int)
```

```python
sns.histplot( data=df[(df["Año"] == 2004) | (df["Año"] == 2014)], x='Cierre', bins = 15, color="lightgreen", hue="Año", element="step", palette=["lightgreen", "lightblue"])
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
plt.title('Distribución de los precios de cierre de la Soja')
```
