# Importancia de las features

Nos piden graficar la importancia de las features. Notemos que al usar One Hot Encoding y Binary Encoding se agregan tantas columnas como valores posibles tiene cada feature.

Al obtener los coeficientes de importancia utilizando ***model.coef_*** obtenemos un cierto coeficiente para cada columna, pero sabemos que en conjunto estas columnas representan una unica feature.

Sabemos que el valor de importance de un feature encodeado se puede aproximar como la suma de los valores absolutos de los features resultantes del encoding.

Luego, tenemos que sumar los valores absolutos de todas las que parten de un mismo, ya que por si solas las columnas no dicen nada. Es decir, sumamos el modulo del coeficiente de importancia de las columnas que salieron de cada feature.

### Sumamos el modulo de los coeficientes de importancia

Sabemos que las features encodeadas con One Hot Encoding o Binary Encoding se corresponden con aquellas columnas que en su nombre terminan con "_0", "_1", "_2", etc.

Luego, para sumar el modulo del coeficiente de importancia de las columnas que salieron de cada feature debemos tener en cuenta lo siguiente:
* Si al final no termina con un numero, entonces no le aplicamos OHE o Binary Encoding.
* Si al final empieza con un "0", entonces comenzamos a sumar.
* Si la siguiente columna termina con una letra o termina con un "0" entonces no debemos seguir sumando y pasamos a la siguiente.

```python
def aplicamos_encoding(feature):
  return (feature[len(feature) - 1].isnumeric())


def feature_actual_terminada(feature_siguiente):
  return (feature_siguiente[len(feature_siguiente) - 1].isalpha()) or ((feature_siguiente[len(feature_siguiente) - 1] == "0"))

def obtener_feature_importance(columnas, coeficientes):
  resultado = []
  suma = 0

  for i in range(len(columnas)):

    if not aplicamos_encoding(columnas[i]):
      resultado.append(np.abs(coeficientes[i]))

    else:

      if (columnas[i][len(columnas[i]) - 1] == "0"):
        suma = 0

      if (i == len(columnas)-1):
        suma = suma + np.abs(coeficientes[i])
        resultado.append(suma)
        continue

      if feature_actual_terminada(columnas[i+1]):
        suma = suma + np.abs(coeficientes[i])
        resultado.append(suma)
        continue

      suma = suma + np.abs(coeficientes[i])

  return resultado
```

```python
feature_importance = obtener_feature_importance(X_train.columns, model.coef_[0])
```
